```python
def jumpingOnClouds(c):
    jumps = 0
    position = 0
    
    while position < len(c) - 1:
        # Attempt to jump by 2. If not possible, jump by 1.
        if position + 2 < len(c) and c[position + 2] == 0:
            position += 2
        else:
            position += 1
        jumps += 1
    
    return jumps

# Sample Input
n = 7
c = [0, 0, 1, 0, 0,1, 0]

# Call the function and print the result
minimum_jumps = jumpingOnClouds(c)
minimum_jumps
```

```python
class Solution(object):
    def isRobotBounded(self, instructions):
                # Directions are represented as vectors: north, east, south, west
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        # Initial position and direction
        x, y = 0, 0
        dir_index = 0  # Start facing north
        
        for instruction in instructions:
            if instruction == 'G':
                # Move one step in the current direction
                x += directions[dir_index][0]
                y += directions[dir_index][1]
            elif instruction == 'L':
                # Turn left: counter-clockwise rotation in the directions array
                dir_index = (dir_index - 1) % 4
            elif instruction == 'R':
                # Turn right: clockwise rotation in the directions array
                dir_index = (dir_index + 1) % 4
                
        # The robot is bounded in a circle if it returns to the origin or is not facing north
        return (x == 0 and y == 0) or dir_index != 0
```

```python
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        count = 0
        # Iterate through each combination of i, j, k
        for i in range(n):
            for j in range(i+1, n):
                for k in range(j+1, n):
                    # Check if the current combination forms an ascending or descending team
                    if (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]):
                        count += 1
        return count
```

Minimize the Maximum of Two Arrays

```python
class Solution(object):
    def minimizeSet(self, divisor1, divisor2, uniqueCnt1, uniqueCnt2):
        def gcd(a, b):
    # Euclidean algorithm for finding GCD
            while b != 0:
                a, b = b, a % b
            return a

        def lcm(a, b):
            # Calculate LCM using the relationship between LCM and GCD
            return abs(a*b) // gcd(a, b)
        
        divisorLcm = lcm(divisor1, divisor2)
        l = 0
        r = 2**31 - 1

        # True if we can take uniqueCnt1 integers from [1..m] to arr1 and take
        # uniqueCnt2 integers from [1..m] to arr2.
        def isPossible(m):
            cnt1 = m - m // divisor1
            cnt2 = m - m // divisor2
            totalCnt = m - m // divisorLcm
            return cnt1 >= uniqueCnt1 and cnt2 >= uniqueCnt2 and \
                totalCnt >= uniqueCnt1 + uniqueCnt2

        while l < r:
            m = (l + r) // 2
            if isPossible(m):
                r = m
            else:
                l = m + 1

        return l
```

```python
class Solution(object):
    def highFive(self, items):
        scores = {}
    
        # Collect all scores for each ID
        for id, score in items:
            if id not in scores:
                scores[id] = []
            scores[id].append(score)
        
        # Calculate top five average for each student
        result = []
        for id in sorted(scores.keys()):
            top_five_avg = sum(sorted(scores[id], reverse=True)[:5]) // 5
            result.append([id, top_five_avg])
        
        return result
        """
        :type items: List[List[int]]
        :rtype: List[List[int]]
        """
        
```

 Trapping Rain Water

```python
class Solution(object):
    def trap(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        if not height:
            return 0
        
        left, right = 0, len(height) - 1  # Two pointers
        leftMax, rightMax = height[left], height[right]  # Max heights to the left and right
        trappedWater = 0  # Result variable
        
        while left < right:
            if leftMax < rightMax:
                left += 1
                leftMax = max(leftMax, height[left])
                trappedWater += leftMax - height[left]
            else:
                right -= 1
                rightMax = max(rightMax, height[right])
                trappedWater += rightMax - height[right]
        
        return trappedWater
```

            return "0"

            result.append("-")

        numerator, denominator = abs(numerator), abs(denominator)
        
        # Append the integer part
        result.append(str(numerator // denominator))
        remainder = numerator % denominator
        
        if remainder == 0:
            return ''.join(result)

        remainders = {}
        
        while remainder:
            if remainder in remainders:
                result.insert(remainders[remainder], "(")

                break
            remainders[remainder] = len(result)
            remainder *= 10
            result.append(str(remainder // denominator))
            remainder %= denominator
        
        return ''.join(result)

Backspace String Compare

```python
class Solution(object):
    def backspaceCompare(self, s, t):
        def buildFinalString(inputString):
            stack = []  # Initialize an empty stack
            for char in inputString:
                if char != '#':
                    stack.append(char)  # Push the character onto the stack
                elif stack:
                    stack.pop()  # Pop the top character if the stack is not empty
            return "".join(stack)  # Join the stack elements to form the final string
        
        # Build the final string for both s and t, and compare them
        return buildFinalString(s) == buildFinalString(t)
        
```

 Reaching Points

```python
class Solution(object):
    def reachingPoints(self, sx, sy, tx, ty):
        # Work backwards from the target to the start
        while tx >= sx and ty >= sy:
            # If we've reached the start point, return True
            if tx == sx and ty == sy:
                return True
            # If tx is greater than ty, we must have gotten here from (tx - ty, ty)
            if tx > ty:
                if ty > sy:
                    tx %= ty
                else:
                    return (tx - sx) % ty == 0
            # If ty is greater than or equal to tx, we must have gotten here from (tx, ty - tx)
            else:
                if tx > sx:
                    ty %= tx
                else:
                    return (ty - sy) % tx == 0
        # If we exit the loop, we couldn't reach the start point
        return False
```

```python
def mostFrequentIPs(logs):
    # Initialize a dictionary to store IP counts
    ipCounts = {}
    
    # Process each log entry
    for log in logs:
        # Split to extract the IP address
        ip = log.split(" - ")[0]
        print(ip,'ip')
        
        # Increment the count for this IP
        if ip in ipCounts:
            ipCounts[ip] += 1
        else:
            ipCounts[ip] = 1
    
    # Find the maximum count
    maxCount = max(ipCounts.values())
    
    # Find all IPs with the maximum count
    mostFrequent = [ip for ip, count in ipCounts.items() if count == maxCount]
    
    return mostFrequent

# Example log entries
logs = [
    "10.0.0.1 - log entry 1 11",
    "10.0.0.1 - log entry 2 213",
    "10.0.0.2 - log entry 1 11",
    "10.0.0.2 - log entry 2 213",
    "10.0.0.2 - log entry 133132"
]

print(mostFrequentIPs(logs))
```

```python
from collections import deque

class Station:
    def __init__(self, name):
        self.name = name
        self.adjacent = []  # List of adjacent stations

    def add_connection(self, station):
        self.adjacent.append(station)

class RailNetwork:
    def __init__(self):
        self.stations = {}  # Maps station names to Station objects

    def add_station(self, name):
        if name not in self.stations:
            self.stations[name] = Station(name)

    def connect_stations(self, name1, name2):
        if name1 in self.stations and name2 in self.stations:
            self.stations[name1].add_connection(self.stations[name2])
            self.stations[name2].add_connection(self.stations[name1])

    def shortestPath(self, fromStationName, toStationName):
        if fromStationName not in self.stations or toStationName not in self.stations:
            return "One or both stations do not exist."
        
        queue = deque([self.stations[fromStationName]])
        visited = {fromStationName: None}  # Track visited stations and their predecessors

        while queue:
            current_station = queue.popleft()

            if current_station.name == toStationName:
                # Build the path from toStationName back to fromStationName
                path = []
                while current_station:
                    path.append(current_station.name)
                    current_station = visited[current_station.name]
                return path[::-1]  # Return reversed path

            for adjacent_station in current_station.adjacent:
                if adjacent_station.name not in visited:
                    queue.append(adjacent_station)
                    visited[adjacent_station.name] = current_station

        return "No path found."

# Example usage
railNetwork = RailNetwork()
railNetwork.add_station("A")
railNetwork.add_station("B")
railNetwork.add_station("C")
railNetwork.add_station("D")
railNetwork.connect_stations("A", "B")
railNetwork.connect_stations("B", "C")
railNetwork.connect_stations("C", "D")

print(railNetwork.shortestPath("A", "D"))  # Output: ['A', 'B', 'C', 'D']
```

```python
def freqQuery(queries):
    element_freq = {}  # Maps elements to their frequencies
    freq_count = {}    # Maps frequencies to their counts
    output = []        # Stores the output for type 3 queries

    for op, value in queries:
        if op == 1:
            # Increment the frequency of the value
            old_freq = element_freq.get(value, 0)
            new_freq = old_freq + 1
            element_freq[value] = new_freq

            # Update the freq_count dictionary
            freq_count[old_freq] = freq_count.get(old_freq, 1) - 1
            freq_count[new_freq] = freq_count.get(new_freq, 0) + 1

        elif op == 2 and value in element_freq and element_freq[value] > 0:
            # Decrement the frequency of the value
            old_freq = element_freq[value]
            new_freq = old_freq - 1
            element_freq[value] = new_freq

            # Update the freq_count dictionary
            freq_count[old_freq] = freq_count.get(old_freq, 1) - 1
            freq_count[new_freq] = freq_count.get(new_freq, 0) + 1

        elif op == 3:
            # Check if any integer is present whose frequency is exactly value
            output.append(1 if freq_count.get(value, 0) > 0 else 0)

    return output

# Example usage
queries = [
    (1, 1),
    (2, 2),
    (3, 2),
    (1, 1),
    (1, 1),
    (2, 1),
    (3, 2)
]
print(freqQuery(queries))  # Output: [0, 1]
```

```python
import math
import os
import random
import re
import sys

#
# Complete the 'countInversions' function below.
#
# The function is expected to return a LONG_INTEGER.
# The function accepts INTEGER_ARRAY arr as parameter.
#

def countInversions(arr):
    def mergeAndCount(arr, temp, left, mid, right):
        i, j, k = left, mid + 1, left
        inv_count = 0

        while i <= mid and j <= right:
            if arr[i] <= arr[j]:
                temp[k] = arr[i]
                i += 1
            else:
                # Since arr[i] > arr[j] and elements are sorted,
                # all elements from arr[i] to arr[mid] will form inversions with arr[j]
                temp[k] = arr[j]
                inv_count += (mid - i + 1)
                j += 1
            k += 1

        # Copy the remaining elements
        while i <= mid:
            temp[k] = arr[i]
            i += 1
            k += 1
        while j <= right:
            temp[k] = arr[j]
            j += 1
            k += 1

        # Copy back to the original array to reflect sorted order
        for i in range(left, right + 1):
            arr[i] = temp[i]

        return inv_count

    # Recursive function to divide the array into halves and merge them while counting inversions.
    def mergeSortAndCount(arr, temp, left, right):
        inv_count = 0
        if left < right:
            mid = (left + right) // 2
            inv_count += mergeSortAndCount(arr, temp, left, mid)
            inv_count += mergeSortAndCount(arr, temp, mid + 1, right)
            inv_count += mergeAndCount(arr, temp, left, mid, right)
        return inv_count

    # Initialize temporary array and start merge sort process
    n = len(arr)
    temp = [0]*n
    return mergeSortAndCount(arr, temp, 0, n - 1)
```

Longest Common Subsequence (LCS) 

```python
def commonChild(s1, s2):
    # Initialize a 2D list (matrix) with zeros
    n = len(s1)
    dp = [[0] * (n + 1) for _ in range(n + 1)]

    # Populate the matrix
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                # If characters match, take the diagonal value and add 1
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                # If not, take the maximum of the left and top values
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    # The value in the bottom-right corner is the length of the longest common subsequence
    return dp[n][n]

# Example usage
s1 = input()
s2 = input()
print(commonChild(s1, s2))
```

```python
def luckBalance(k, contests):
    # Separate important and unimportant contests
    important_contests = [luck for luck, importance in contests if importance == 1]

    unimportant_contests = [luck for luck, importance in contests if importance == 0]

    
    # Sort important contests in descending order of luck
    important_contests.sort(reverse=True)
    print(important_contests)
    
    # Sum of luck from losing all unimportant contests
    luck_balance = sum(unimportant_contests)
    
    # Lose k highest luck important contests, win the rest
    luck_balance += sum(important_contests[:k])  # Sum of luck from losing allowed important contests
    luck_balance -= sum(important_contests[k:])  # Subtract luck from winning the rest of the important contests
    
    return luck_balance

# Example usage
k = 2
contests = [
    [5, 1],
    [1, 1],
    [4, 0]
]

print(luckBalance(k, contests))
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    # Initialize the dummy head of the result list.
    dummy_head = ListNode(0)
    current = dummy_head
    carry = 0

    # Loop through lists l1 and l2 until you reach both ends.
    while l1 or l2:
        x = l1.val if l1 else 0
        y = l2.val if l2 else 0
        sum = carry + x + y
        carry = sum // 10
        current.next = ListNode(sum % 10)
        current = current.next
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    # If the final sum has a carry over, add a new node with carry value.
    if carry > 0:
        current.next = ListNode(carry)

    # Return the dummy head's next node, which is the start of the result list.
    return dummy_head.next

# Helper function to convert list to ListNode
def list_to_listnode(lst):
    dummy_root = ListNode(0)
    ptr = dummy_root
    for number in lst:
        ptr.next = ListNode(number)
        ptr = ptr.next
    ptr = dummy_root.next
    return ptr

# Helper function to convert ListNode to list
def listnode_to_list(node):
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result

# Example 1
l1 = list_to_listnode([2,4,3])
l2 = list_to_listnode([5,6,4])
result1 = addTwoNumbers(l1, l2)

# Example 2
l1_example2 = list_to_listnode([0])
l2_example2 = list_to_listnode([0])
result2 = addTwoNumbers(l1_example2, l2_example2)

# Example 3
l1_example3 = list_to_listnode([9,9,9,9,9,9,9])
l2_example3 = list_to_listnode([9,9,9,9])
result3 = addTwoNumbers(l1_example3, l2_example3)

listnode_to_list(result1), listnode_to_list(result2), listnode_to_list(result3)
```
